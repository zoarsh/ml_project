```python
import pandas as pd
from pathlib import Path

# --- 1) הגדרות ---
key_col   = "mmaanecode"                  # מפתח המיזוג
name_cols = ["שםתוכנית", "שםתוכנית_A"]   # שמות התוכנית (אם קיימים)
attrs = [
    "יחידהאירגונית","היכןמתקיימתהתוכנית",
    "תפוקותדרךפעולה","תפוקותאוכלוסייתיעד","תפוקותהרכבדרךפעולה",
    "תפוקותתדירותהפעילות","תפוקותאורךהפעילות",
    "תוצאהעיקריתרצויה","תוצאהמשניתרצויה",
    "טיפולפיזי","השתייכותלמשפחה","הגנהמפניאחרים","לימודי",
    "השתייכותוהשתלבותחברתית","רווחהרגשית","הגנהמפניהתנהגויותסיכון"
]

# --- 2) פונקציית אגירה: אם כל הערכים זהים → יחיד; אחרת → ייחודיים מחוברים בנקודה-פסיק ---
def collapse_values(s: pd.Series):
    s = s.dropna().astype("string").str.strip()
    if s.empty:
        return pd.NA
    uniq = pd.unique(s)
    return uniq[0] if len(uniq) == 1 else "; ".join(sorted(uniq))

# --- 3) בחירת עמודות שקיימות בפועל ---
all_wanted = [key_col] + [c for c in name_cols if c in df_prog.columns] + [c for c in attrs if c in df_prog.columns]
dfp = df_prog[all_wanted].copy()

# --- 4) בניית טבלת מאסטר: שורה אחת לכל קוד תוכנית ---
agg_dict = {c: collapse_values for c in all_wanted if c != key_col}
program_master = (
    dfp.groupby(key_col, dropna=False, as_index=False)
       .agg(agg_dict)
)

# --- 5) סדר עמודות: קוד → שמות → מאפיינים ---
ordered_cols = [key_col] + [c for c in name_cols if c in program_master.columns] + [c for c in attrs if c in program_master.columns]
program_master = program_master[ordered_cols]

# --- 6) בדיקות קצרות + הצצה ---
n_codes = program_master.shape[0]
print(f"מספר שורות (קודי תוכנית ייחודיים): {n_codes:,}")


# --- 7) שמירה לקבצים ---
out = Path("data/external/programs_master_by_code.pkl")
program_master.to_pickle(out)
program_master.to_csv(out.with_suffix(".csv"), index=False, encoding="utf-8-sig")
print("✅ Saved:", out.resolve())
```

```python
#השוואת דמיון בין שמות התוכניות בקובץ שלנו לתוכניות בקובץ החיצוני
import re, difflib, pandas as pd
from pathlib import Path

def normalize(s):
    if s is None: return ""
    s = str(s)
    s = s.replace("״", '"').replace("”","\"").replace("’","'").replace("–","-").replace("־","-")
    s = re.sub(r"[^\w\u0590-\u05FF\s-]", " ", s)   # השאר עברית, ספרות, קו-אמצעי ורווח
    s = re.sub(r"\s+", " ", s).strip()
    # token sort
    toks = s.split()
    toks.sort()
    return " ".join(toks)

def similarity(a,b):
    return int(round(difflib.SequenceMatcher(None, normalize(a), normalize(b)).ratio()*100))

# 1) טוענים שמות מהדאטה
df_main = pd.read_pickle(Path("data/interim/data_cleansing.pkl"))
prog_names = df_main["program_name_m1_clean"].dropna().astype("string").str.strip().unique().tolist()
serv_names = df_main["service_name_m1_clean"].dropna().astype("string").str.strip().unique().tolist()
all_names = sorted(set(prog_names + serv_names))

# 2) 12 התוכניות מהמאסטר
missing_prog_names = [
    'מרכז חכ"מ – מרכז לחיזוק כוחות משפחתיים - מרכז יום טיפולי לילדים גילאי 5-13',
    "פעילות קיץ הפגתית לילדים בסיכון",
    "טיפול בילדים נפגעי אלימות במשפחתם וטיפול בהוריהם האלימים",
    "התערבויות דרך שטח לנוער בסיכון ובמצבי סיכון (קידום נוער)",
    "עצמה לכל - תכנית חונכות וקריאה",
    "תעשייחד",
    "מצעד הספרים לחיזוק הקשר בין הורים וילדים",
    'מרח"ב - מענה רווחתי חינוכי בריאותי בבית הספר',
    "שערי הישיבה",
    'להב"ה - למען נוער בסיכון במרחב הכפרי (משרד הרווחה)',
    "מועדונית שיקומית א' (3ימים)",
    "התוכנית לקולנוע מסייע – חצי שנתית",
]

# 3) מוצאים התאמה הטובה ביותר לכל שם חסר
rows = []
for name in missing_prog_names:
    scores = [(cand, similarity(name, cand)) for cand in all_names]
    best = max(scores, key=lambda x: x[1]) if scores else ("", 0)
    rows.append((name, best[0], best[1]))

pd.DataFrame(rows, columns=["תוכנית חסרה (מאסטר)", "הקרובה ביותר בדאטה", "ציון דמיון (0-100)"])
```

```python
import pandas as pd
from pathlib import Path


df = pd.read_pickle(Path("data/interim/data_cleansing_merged_try.pkl"))

# -------- 1) דגל: program_info_missing --------
prog_cols = [c for c in df.columns if str(c).startswith((
    "program_", "outputs_", "desired_", "physical_", "family_",
    "protection_", "educational_", "social_", "emotional_"
))]

def is_missing_like_frame(frame: pd.DataFrame) -> pd.DataFrame:
    S = frame.astype("string").apply(lambda s: s.str.strip().str.lower())
    return S.isna() | S.eq("") | S.eq("0") | S.eq("unknown") | S.eq("לא ידוע")

if prog_cols:
    miss_like = is_missing_like_frame(df[prog_cols])
    df["program_info_missing"] = miss_like.all(axis=1).astype(int)
else:
    df["program_info_missing"] = 0

print("program_info_missing counts:\n", df["program_info_missing"].value_counts())
print("\nprogram_info_missing %:\n", df["program_info_missing"].value_counts(normalize=True).mul(100).round(2))

# -------- 2) המרת 'לא ידוע' -> 'unknown' בעמודות רקע --------
def classify(col: str) -> str:
    s = str(col).lower()
    if s.startswith("rs"): return "rs"
    if s.startswith("ri"): return "ri"
    if s.startswith("st"): return "st"
    return "other"

cols_rs = [c for c in df.columns if classify(c)=="rs"]
cols_ri = [c for c in df.columns if classify(c)=="ri"]
cols_st = [c for c in df.columns if classify(c)=="st"]

background_cols = [c for c in df.columns
                   if (c not in prog_cols) and (c not in cols_rs) and (c not in cols_ri) and (c not in cols_st)]

# ננקה רק עמודות טקסט/קטגוריה, לא נוגעים במספרים/תאריכים
bg_to_clean = [c for c in background_cols
               if (df[c].dtype == "object") or str(df[c].dtype).startswith(("category","string"))]

replaced_total = 0
for c in bg_to_clean:
    s = df[c]
    # אם קטגוריאלי – נוסיף את 'unknown' לרשימת הקטגוריות לפני ההחלפה
    if pd.api.types.is_categorical_dtype(s):
        if "unknown" not in s.cat.categories:
            s = s.cat.add_categories(["unknown"])
        mask = s.astype("string").str.strip().isin(["לא ידוע", "לא ידוע - בבדיקה"])
        replaced_total += int(mask.sum())
        df[c] = s.where(~mask, other="unknown")
    else:
        s_str = s.astype("string")
        mask = s_str.str.strip().isin(["לא ידוע", "לא ידוע - בבדיקה"])
        replaced_total += int(mask.sum())
        df.loc[mask, c] = "unknown"

print(f"\nreplaced 'לא ידוע' -> 'unknown' total: {replaced_total:,}")

# --- בדיקת sanity קצרה: האם כעת הדגל מזהה ~15.8%? ---
print("\npost-check program_info_missing %:")
print(df["program_info_missing"].value_counts(normalize=True).mul(100).round(2))
```

```python
#דוח שינויים בקובץ

import pandas as pd
from pathlib import Path

# ---------- נתיבים ----------
BASE = Path("data/interim")
MAIN_IN = BASE / "data_cleansing.pkl"              # בסיס לפני מיזוג
OUT_TRY = BASE / "data_cleansing_merged_try.pkl"   # אחרי מיזוג (ניסיוני)
OUT_CLEAN = BASE / "data_cleansing_merged_clean_ready.pkl"

# ---------- טעינה ----------
df_base = pd.read_pickle(MAIN_IN)
df = pd.read_pickle(OUT_TRY)
print("before:", df_base.shape, "| after:", df.shape)

# ---------- עמודות תוכנית (שנוספו מהמיפוי) ----------
prog_cols = [c for c in df.columns if str(c).startswith((
    "program_", "outputs_", "desired_", "physical_", "family_",
    "protection_", "educational_", "social_", "emotional_"
))]
added_cols = [c for c in df.columns if c not in df_base.columns]
added_prog_cols = [c for c in added_cols if c in prog_cols]
print(f"\nadded program columns: {len(added_prog_cols)}")
for c in added_prog_cols:
    print(" •", c)

# ---------- דגל: אין מידע תוכנית (כל עמודות התוכנית ריקות/NaN/'0'/unknown) ----------
def missing_like(s):
    s = s.astype("string").str.strip().str.lower()
    return s.isna() | s.eq("") | s.eq("0") | s.eq("unknown") | s.eq("לא ידוע")

if "program_info_missing" not in df.columns:
    mask_no_info = df[prog_cols].apply(missing_like).all(axis=1) if prog_cols else pd.Series(False, index=df.index)
    df["program_info_missing"] = mask_no_info.astype(int)

print("\nprogram_info_missing counts:")
print(df["program_info_missing"].value_counts(dropna=False))
print(df["program_info_missing"].value_counts(normalize=True).mul(100).round(2))

# ---------- המרת 'לא ידוע'→'unknown' בעמודות רקע (לא rs/ri/st ולא עמודות תוכנית) ----------
def classify(col: str) -> str:
    s = str(col).lower()
    if s.startswith("rs"): return "rs"
    if s.startswith("ri"): return "ri"
    if s.startswith("st"): return "st"
    return "other"

cols_rs = [c for c in df.columns if classify(c)=="rs"]
cols_ri = [c for c in df.columns if classify(c)=="ri"]
cols_st = [c for c in df.columns if classify(c)=="st"]

background_cols = [c for c in df.columns
                   if (c not in prog_cols) and (c not in cols_rs) and (c not in cols_ri) and (c not in cols_st)]

bg_to_clean = [c for c in background_cols
               if df[c].dtype == "object" or str(df[c].dtype).startswith(("category","string"))]

replacements = {}
for c in bg_to_clean:
    s = df[c].astype("string")
    mask = s.str.strip().isin(["לא ידוע", "לא ידוע - בבדיקה"])
    n = int(mask.sum())
    if n > 0:
        df.loc[mask, c] = "unknown"
    replacements[c] = n

total_replaced = int(sum(replacements.values()))
nonzero_rep = {k:v for k,v in replacements.items() if v>0}

# ---------- סיכום ----------
summary = {
    "rows_after": len(df),
    "cols_before": df_base.shape[1],
    "cols_after": df.shape[1],
    "added_cols_total": len(added_cols),
    "added_program_cols": len(added_prog_cols),
    "program_info_missing_1": int((df['program_info_missing']==1).sum()),
    "program_info_missing_1_pct": round(100*(df['program_info_missing']==1).mean(), 2),
    "replaced_to_unknown_total": total_replaced,
}
print("\nSUMMARY")
for k,v in summary.items():
    print(f"- {k}: {v}")

if nonzero_rep:
    print("\nTop replacements (unknown):")
    for k,v in sorted(nonzero_rep.items(), key=lambda x: x[1], reverse=True)[:10]:
        print(f" • {k}: {v}")

# ---------- שמירות ----------
df.to_pickle(OUT_CLEAN)
(pd.Series(added_prog_cols, name="added_program_columns")
   .to_frame()
   .to_csv(BASE / "added_program_columns.csv", index=False, encoding="utf-8-sig"))
pd.DataFrame([summary]).to_csv(BASE / "merge_clean_summary.csv", index=False, encoding="utf-8-sig")

print("\nSaved:")
print(" •", OUT_CLEAN.name)
print(" • added_program_columns.csv")
print(" • merge_clean_summary.csv")
```
