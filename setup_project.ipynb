```python
#סידור הקבצים בתיקיות
import shutil
from pathlib import Path

# נתיב בסיס (התיקייה שבה הקבצים שלך נמצאים עכשיו)
base = Path(".")

# יצירת מבנה תיקיות
(project := base / "my_ml_project").mkdir(exist_ok=True)
for sub in ["notebooks", "data/raw", "data/interim", "docs/codebook", "docs/sps", "docs/lessons"]:
    (project / sub).mkdir(parents=True, exist_ok=True)

# העברות לפי החלטה
moves = {
    "data_prep.ipynb": project / "notebooks" / "data_prep.ipynb",
    "participants_sep23_sep24.csv": project / "data/interim" / "participants_sep23_sep24.csv",
    "raw_data/participants_sep23_sep24_raw.csv": project / "data/raw" / "participants_sep23_sep24_raw.csv",
    "ML Project Structure.pdf": project / "docs" / "ML Project Structure.pdf",
    "codebook_template.xlsx": project / "docs/codebook" / "codebook_template.xlsx",
    "variable_name_mapping.xlsx": project / "docs/codebook" / "variable_name_mapping.xlsx",
    "עותק codebook 360.xlsx": project / "docs/codebook" / "עותק codebook 360.xlsx",
    "עותק codebook 360.pdf": project / "docs/codebook" / "עותק codebook 360.pdf",
    "בניה מחדש לשתי מדידות.sps.txt": project / "docs/sps" / "בניה מחדש לשתי מדידות.sps.txt",
    "סינטקס סידור משתנים מקובץ משתתפים.sps- תיקונים זהר 03.06.25.sps.txt": project / "docs/sps" / "סינטקס סידור משתנים.sps.txt",
}

# הפעלת ההעברה
for src, dst in moves.items():
    src_path = base / src
    if src_path.exists():
        shutil.move(str(src_path), str(dst))
        print(f"Moved {src_path} → {dst}")
    else:
        print(f"Skipped (not found): {src}")

```

```python
from pathlib import Path

# נניח שמריצים מתוך notebooks/, השורש הוא התיקייה מעל
ROOT = Path.cwd().parent
print("Project root:", ROOT)

gi = ROOT / ".gitignore"

# מה כדאי להתעלם (אפשר לשנות/להוסיף לפי הצורך)
entries = [
    "# Project ignores",
    "artifacts/",                 # snapshots
    "reports/",                   # דוחות HTML/PDF אם תשמרי לשם
    "notebooks/*.html",           # דוחות Sweetviz/EDA במחברת
    "*.ipynb_checkpoints/",
    "__pycache__/",
    ".DS_Store",
    "data/raw/",                  # נתונים גולמיים כבדים/רגישים
    "data/interim/",              # קבצי עבודה ביניים
    # הוסיפי כאן עוד אם תרצי
]

existing = gi.read_text(encoding="utf-8", errors="ignore") if gi.exists() else ""
to_add = [ln for ln in entries if ln not in existing]

if to_add:
    prefix = "" if (existing.endswith("\n") or existing == "") else "\n"
    gi.write_text(existing + prefix + "\n" + "\n".join(to_add) + "\n", encoding="utf-8")
    print("Updated .gitignore at:", gi)
else:
    print(".gitignore already up to date at:", gi)
```

```python
# Ensure artifacts/ is ignored by Git (idempotent).
def ensure_gitignore_has_artifacts():
    """
    Rationale: Snapshots shouldn't go to Git.
    Action: Append 'artifacts/' to .gitignore if missing, or create file.
    """
    gi = Path(".gitignore")
    line = "artifacts/\n"
    if gi.exists():
        text = gi.read_text(encoding="utf-8", errors="ignore")
        if "artifacts/" not in text:
            gi.write_text(text.rstrip() + "\n\n# Data snapshots\n" + line, encoding="utf-8")
            print("[.gitignore] added 'artifacts/'")
        else:
            print("[.gitignore] already contains 'artifacts/'")
    else:
        gi.write_text("# Data snapshots\n" + line, encoding="utf-8")
        print("[.gitignore] created with 'artifacts/'")
```
